Load the Dataset

```python
import pandas as pd
import numpy as np

data = pd.read_csv('./basic-details-of-schools.csv')
```

Handle Missing Data

A. Identify Columns with Missing Data

```python
missing_data = data.isnull().sum().sort_values(ascending=False)
missing_data
```

```python
missing_percentage = (missing_data / len(data)) * 100
missing_percentage
```

```python
missing_df = pd.DataFrame({'Missing Values': missing_data, 'Percentage':missing_percentage})
missing_df
```

B. Drop Columns with Excessive Missing Data

```python
columns_to_drop = ['aff_board_h_sec','ward','aff_board_sec','pincode','district_code','subdistrict_code']
data.drop(columns = columns_to_drop, inplace = True)
```

C. Handle Missing Values in Remaining Columns

Categorical Columns:

use "Unknown" for subdistrict_name.

```python
categorical_cols = ['subdistrict_name']
for col in categorical_cols:
    data[col].fillna('Unknown', inplace=True)
```

Analyzing the data again (new missing data)

```python
new_missing_data = data.isnull().sum().sort_values(ascending=False)
new_missing_data
```

```python
data.info()
```

Removing Rows with in which location_type and village_name were empty(' ')

Also rows were only 1 and 1 for location_type and village_name resp.

```python
data = data.dropna(subset=['location_type', 'village_name'])
```

Convert Data Types

A. Floats to Integers

```python
float_cols_to_int = ['class_from', 'class_to', 'pre_primary', 'class_rooms',
       'other_rooms', 'total_teachers', 'pre_primary_students', 'i_students',
       'ii_students', 'iii_students', 'iv_students', 'v_students',
       'vi_students', 'vii_students', 'viii_students', 'ix_students',
       'x_students', 'xi_students', 'xii_students', 'class_students',
       'class_with_pre_primary_students']
data[float_cols_to_int] = data[float_cols_to_int].astype('Int64')
```

```python
data.columns
```

```python
data.info()
```

B. Ensure Proper Data Types for ID Columns

Convert ID columns to str

```python
id_cols = ['id','state_code', 'udise_village_code', 'udise_school_code']
data[id_cols] = data[id_cols].astype(str)
```

C. Datetime64[ns] to Int64

```python
data['year_of_establishment'] = pd.to_numeric(data['year_of_establishment'], errors='coerce').astype('int64')
```

Duplicate Handling

Find columns with any duplicates

```python
duplicate_columns = [col for col in data.columns if data[col].duplicated().any()]
print("Columns with duplicates:", duplicate_columns)
```

Check for duplicates in the dataset, particularly in key columns like udise_school_code.

```python
# Check for duplicates
duplicates = data.duplicated(subset=['udise_school_code'], keep=False)
print(data[duplicates])

# Drop duplicates
data.drop_duplicates(subset=['udise_school_code'], inplace=True)
```

Validate Coordinates

Remove or flag rows with invalid longitude or latitude.

```python
data = data[(data['longitude'] >= -180) & (data['longitude'] <= 180)]
data = data[(data['latitude'] >= -90) & (data['latitude'] <= 90)]
```

```python
data.columns
```

```python
#The error 'index' object is not callable typically occurs in Python when you're trying to call (use parentheses ()) on an object named index that is not a callable function or method.
```

Standardize Text Data

Clean text columns like state_name, district_name, etc., for consistency.

```python
text_cols = ['id', 'state_name', 'state_code', 'district_name', 'subdistrict_name',
       'cluster_name', 'village_name', 'udise_village_code', 'school_name',
       'udise_school_code', 'school_category', 'school_type', 'management']
for col in text_cols:
    data[col] = data[col].str.strip().str.title()
```
